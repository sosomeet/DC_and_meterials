```python
import pandas as pd
import plotly.graph_objects as go

# ----------------------------
# 1) Load
# ----------------------------
df = pd.read_csv("./merged_spillover_data.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

num_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()
num_cols = num_cols[:5]   # 5 

print("Using columns:", num_cols)

# ----------------------------
# 2)  (  )
# ----------------------------
df_std = df.copy()
for col in num_cols:
    df_std[col] = (df[col] - df[col].mean()) / df[col].std()

# ----------------------------
# 3) Plotly  
# ----------------------------
fig = go.Figure()

for col in num_cols:
    fig.add_trace(
        go.Scatter(
            x=df_std["Date"],
            y=df_std[col],
            mode="lines",
            name=col
        )
    )

fig.update_layout(
    title="Interactive Overlay (Click legend to toggle)",
    xaxis_title="Date",
    yaxis_title="Z-score",
    height=600,
    hovermode="x unified"
)

fig.show()
```
